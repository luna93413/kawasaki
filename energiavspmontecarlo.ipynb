```python
import matplotlib.pyplot as plt

# Nombre del archivo generado por el programa en C
archivo = "energia_pmontecarlo.txt"

# Listas para almacenar los datos
iteraciones = []
energias = []

# Leer el archivo y extraer los datos
with open(archivo, "r") as file:
    for linea in file:
        energia, iteracion = linea.split()
        energias.append(float(energia))
        iteraciones.append(int(iteracion))

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.plot(iteraciones, energias, marker="o",markersize=2, linestyle=" ", color="b", label="Energa vs Iteraciones")

# Configurar etiquetas y ttulo
plt.xlabel("Pasos montecarlo", fontsize=12)
plt.ylabel("Energa", fontsize=12)
plt.title("Energa vs Pasos montecarlo", fontsize=14)
plt.legend()
plt.grid()

# Mostrar la grfica
plt.show()
```
