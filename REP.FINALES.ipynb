```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos de los dos ficheros
# Cambia los nombres de los ficheros si es necesario
datos_positivos = np.loadtxt('densidadpositivo.txt')
datos_negativos = np.loadtxt('densidadnegativo.txt')

# Asignar columnas a variables
densidad_pos = datos_positivos[:, 0]
T_pos = datos_positivos[:, 1]

densidad_neg = datos_negativos[:, 0]
T_neg = datos_negativos[:, 1]

# Graficar ambas densidades en funci√≥n de la temperatura
plt.plot(T_pos, densidad_pos, 'o-', label='Densidad positiva')
plt.plot(T_neg, densidad_neg, 's-', label='Densidad negativa')

plt.xlabel('Temperatura (T)')
plt.ylabel('Densidad')
plt.title('Densidad positiva y negativa vs Temperatura')
plt.legend()
plt.show()
```

```python
# ...existing code...
import numpy as np
import matplotlib.pyplot as plt

# Nombres de los ficheros para cada valor de N
fichero_N1 = 'Mag32.txt'
fichero_N2 = 'Mag64.txt'
fichero_N3 = 'Mag128.txt'

# Cargar los datos de cada fichero
datos_N1 = np.loadtxt(fichero_N1)
datos_N2 = np.loadtxt(fichero_N2)
datos_N3 = np.loadtxt(fichero_N3)

# Asignar columnas a variables
mag_N1 = datos_N1[:, 0]
err_N1 = datos_N1[:, 1]
T_N1 = datos_N1[:, 2]

mag_N2 = datos_N2[:, 0]
err_N2 = datos_N2[:, 1]
T_N2 = datos_N2[:, 2]

mag_N3 = datos_N3[:, 0]
err_N3 = datos_N3[:, 1]
T_N3 = datos_N3[:, 2]

# Graficar las curvas de magnetizaci√≥n con barras de error
plt.errorbar(T_N1, mag_N1, yerr=err_N1, fmt='o-', label='N1')
plt.errorbar(T_N2, mag_N2, yerr=err_N2, fmt='s-', label='N2')
plt.errorbar(T_N3, mag_N3, yerr=err_N3, fmt='^-', label='N3')

plt.xlabel('Temperatura (T)')
plt.ylabel('Magnetizaci√≥n')
plt.title('Magnetizaci√≥n vs Temperatura para tres valores de N')
plt.legend()

plt.show()
```
